# Studying the Central Limit Theorem in Statistics

The Central Limit Theorem (CLT) is one of the most important theorems in statistics, allowing the usage of Gaussian distribution approximations even when the initial distribution is non-Gaussian, stating that:

*If a large enough sample is drawn from a population, regardless of the distribution the population is following, the distribution of the sample mean will tend to follow a Gaussian distribution with following properties*:

- The mean value of the sample mean distribution is equal to the initial populations mean value.

- The standard deviation of the sample mean distribution is calle **Standard Error** and is given from the formula $\frac{\sigma_{population}}{\sqrt{n_{sample}}}$.

For the following studies of CLT, the sample mean will be defined as:

- $\bar{x} = \frac{1}{N}\sum_{i=1}^{n} x_{i}$

In order to study the validity of CLT, a distribution composed of random numbers following a $\chi^2$ distribution with one degree of freedom is plotted. Then samples of size 10,20 and 100 are drawn from the initial distriubtion and the distribution of their average is plotted. The number of observations that is used is equal to 10000 to allow for enough statistics for this study.

```python
import numpy as np
import matplotlib.pyplot as plt
import stats_utils as stats
```

```python
rands_chi = np.random.chisquare(1,10000)
stats.plot(rands_chi,"Random Numbers","$\chi_{1}^2$ Distribution",100)
```

As a first test, 10000 samples of size 10 is drawn for the initial $\chi_{1}^2$ and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_chi,10,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
```

From the above figure it can be observed that the distribution of the sample mean is starting to resemble a Gaussian distribution. However, due to the initial skewness of the $\chi^2_{1}$ distribution, the shape cannot be characterized as Gaussian yet, since the size of the sample was not big enough. Next 10000 samples of size 20 is drawn for the initial $\chi_{1}^2$ and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_chi,20,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
```

From the above figure it can be observed that the distribution of the sample mean is looking more like a Gaussian distribution, however some skewness is still persisiting. Finally, 10000 samples of size 100 is drawn for the initial $\chi_{1}^2$ and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_chi,100,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
std_err = (np.std(rands_chi))/(np.sqrt(100))
print(f"Standard Error: {std_err}")
```

**Conclusion**: As it can be observed from the above figure, when the size of the sample drawn for the initial distribution is large enough, the sample mean is following a Gaussian distribution. The mean value of the distribution is very close to the mean value of the initial $\chi^2_{1}$ distribution and the Standard Error follows the formula $\frac{\sigma}{\sqrt{n}}$. 

As a next step, the validity of the CLT was tested for a population composed of random numbers that follow different distributions. The distributions that were used were an exponential distribution with $\lambda = 1$ and a uniform distribution with a lower value of 0 and an upper value of 2. 10000 obervations from both distributions were inserted in the population to prevent one of two distributions dominating the populations. The mean value of the popluation is calculated according to the weighted mean value $\mu_{weighted} = \frac{N_{1}\mu_{1}+N_{2}\mu_{2}}{N_{1} + N_{2}}$.

```python

rands_exp = np.random.exponential(1,10000)
rands_uniform = np.random.uniform(0,2,10000)
rands_mix = np.concatenate((rands_uniform,rands_exp))
stats.plot(rands_mix,"Random Numbers","Merged Uniform and Exponential distribution",100)
```

Initially, 10000 samples of size 10 is drawn for the initial population and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_mix,10,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
```

From the above figure it can be observed that the distribution of the sample mean is starting to resemble a Gaussian distribution. However, due to the initial skewness of the initial population, the shape cannot be characterized as Gaussian yet, since the size of the sample was not big enough. Next 10000 samples of size 20 is drawn for the initial population and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_mix,20,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
```

From the above figure it can be observed that the distribution of the sample mean is looking more like a Gaussian distribution, however some skewness is still persisiting. Finally, 10000 samples of size 100 is drawn for the initial population and the distribution of the sample average is plotted:

```python
stats.get_full_plot(rands_mix,100,"mean",r"$\bar{x}$",r"Sample average $\bar{x}$ distribution",100,10000)
std_err = np.std(rands_mix)/(np.sqrt(100))
print(f"Standard Error: {std_err}")
```

**Conclusion**: As it can be observed from the above figure, when the size of the sample drawn for the initial distribution is large enough, the sample mean is following a Gaussian distribution. The mean value of the distribution is very close to the mean value of the initial distribution and the Standard Error follows the formula $\frac{\sigma}{\sqrt{n}}$. The CLT therefore works for a mixture of random variables too, provided that the distribution of the mixture has a finite mean, finite variance and the observations drawn are independent.
